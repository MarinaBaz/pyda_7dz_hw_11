**Задание 1**

Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
	- оценка 2 и меньше - низкий рейтинг
	- оценка 4 и меньше - средний рейтинг
	- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
film_ratings = pd.read_csv('ratings.csv')
film_ratings.head()
```

```python
def ratings_level(var_rating):
    if var_rating <= 2:
        return 'lower' #низкий рейтинг
    elif var_rating > 2 and var_rating <=4:
        return 'medium' #средний рейтинг
    else:
        return 'high' #высокий рейтинг

film_ratings['level'] = film_ratings['rating'].apply(ratings_level)
film_ratings.head(20)
```

```python
def know_region(str):
    list_str = str.split(' ')
    set_str = set(list_str)
    for key,value in geo_data.items():
        set_geo = set(value)
        if set_str & set_geo == set():
            return 'undefined'
        else:
            return key

data['region'] = data['keyword'].apply(know_region)
data.head()
```
